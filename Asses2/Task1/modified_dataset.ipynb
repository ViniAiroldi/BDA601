```python
# import pandas with shortcut 'pd'
import pandas as pd  
  
# read_csv function which is used to read the required CSV file
data = pd.read_csv('BDA601_Assessment 2_Telco-Customer-Churn_downloaded 05082020.csv')
  
# display 
print("Original 'BDA601_Assessment 2_Telco-Customer-Churn_downloaded 05082020.csv' CSV Data: \n")
print(data)
```

```python
# drop function which is used in removing or deleting rows or columns from the CSV files
data.drop('MonthlyCharges', inplace=True, axis=1)
data.drop('OnlineSecurity', inplace=True, axis=1)
data.drop('StreamingTV', inplace=True, axis=1)
data.drop('InternetService', inplace=True, axis=1)
data.drop('Partner', inplace=True, axis=1)

# display 
print("\nCSV Data after deleting the column 'MonthlyCharges':\n")
print("\nCSV Data after deleting the column 'OnlineSecurity':\n")
print("\nCSV Data after deleting the column 'StreamingTV':\n")
print("\nCSV Data after deleting the column 'InternetService':\n")
print("\nCSV Data after deleting the column 'Partner':\n")
print(data)
```
