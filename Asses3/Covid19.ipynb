```python
# Importing the libraries

import pandas as pd
import numpy as np
import plotly.express as px
```

```python
px
```

```python
# Novel Corona Virus (COVID-19) epidemiological data since 22 January 2020. The data is compiled by the Johns Hopkins University Center for Systems Science and Engineering (JHU CCSE).

base_url = 'https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_confirmed_global.csv&filename=time_series_covid19_confirmed_global.csv'
confirmed_df = pd.read_csv(base_url + 'time_series_covid19_confirmed_global.csv')
```

```python
# Check the data

confirmed_df
```

```python
# Deleting unnecessary data.

confirmed_df = confirmed_df.drop(columns=['Lat', 'Long', 'Province/State'])
```

```python
# Check the data

confirmed_df.head()
```

```python
# Grouping regions from the same country

confirmed_df = confirmed_df.groupby(by='Country/Region').aggregate(np.sum).T
```

```python
# Check the data

confirmed_df
```

```python
# Transforming dates in index

confirmed_df.index.name = 'Date'
confirmed_df = confirmed_df.reset_index()
```

```python
# Check the data

confirmed_df.tail()
```

```python
# Melt dataframe

confirmed_melt_df = confirmed_df.melt(id_vars='Date').copy()
```

```python
# Check again

confirmed_melt_df
```

```python
# Rename columns Value to Confirmed

confirmed_melt_df.rename(columns={'value':'Confirmed'}, inplace=True)
```

```python
# Check

confirmed_melt_df.head()
```

```python
# Correct issues from date/month, show the error

max_date = confirmed_melt_df['Date'].max()
max_date
```

```python
# Change the format

confirmed_melt_df['Date'] = pd.to_datetime(confirmed_melt_df['Date'])
```

```python
# Check

confirmed_melt_df.head()
```

```python
# Change the format to day, month and year

confirmed_melt_df['Date'] = confirmed_melt_df['Date'].dt.strftime('%m/%d/%Y')
```

```python
# Check

max_date = confirmed_melt_df['Date'].max()
max_date
```
